```python
import  pandas as pd
import numpy as np

df = pd.read_csv('netflix1.csv')
df.head()


```

```python
df.shape
print(f"The number of rows in the data are {df.shape[0]} and number of columns are {df.shape[1]}")

```

```python
df.info()
```

```python
df.columns
df.drop(columns="Country",inplace=True)
df.head()   #column of date added removed
```

```python
new_column=[]
for i in df.columns:
    new_column.append(i.capitalize())

df.columns=new_column
df.columns  #to start the first name with capital letter
```

```python
# to see some duplicate
df.duplicated().sum()   # we have no duplicates in our data
```

```python
#to drop duplicates
df.drop_duplicates(inplace=True)
```

```python
# to drop nan values,first checking if there are nan values
df.isna().sum()

```

```python
df.head()
df['Show_id'].unique()
df['Show_id'].apply(lambda x:x.split('s')[1])
  #we can see that the show id is string dtype we should convert it into int
 
```
